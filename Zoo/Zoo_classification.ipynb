```
x=f.drop('class_type',axis=1)
y=f['class_type']
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```
print(x_train)
```

```
x_train=np.array(x_train).reshape(-1,16)
y_train=np.array(y_train).reshape(-1,1)
```

```
print(x_train)
```

```
x_train.shape
```

```
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=4)
clf.fit(x_train,y_train)
```

```
y_pred=clf.predict(x_test)
```

```
from sklearn import metrics
accuracy=metrics.accuracy_score(y_pred,y_test)
print("Accuracy : %.2f"%accuracy)
```

```
from sklearn.metrics import confusion_matrix , classification_report
print("confusion matrix : ",confusion_matrix(y_pred,y_test))
```

```
print(classification_report(y_pred,y_test))
```

```
accuracy=[]
for i in range(1,50):
  clf1=KNeighborsClassifier(n_neighbors=i)
  clf1.fit(x_train,y_train)
  ypred=clf1.predict(x_test)
  accuracy.append(metrics.accuracy_score(ypred,y_test))
```

```
accuracy
```

```
max(accuracy)
```

```
from sklearn.linear_model import LogisticRegression
lf=LogisticRegression()
lf.fit(x_train,y_train)
```

```
Ypred=lf.predict(x_test)
```

```
print("Logistic accuracy : %.2f"%metrics.accuracy_score(Ypred,y_test))
```

```
from sklearn.metrics import confusion_matrix
print("Confusion matrix : ",confusion_matrix(Ypred,y_test))
```

```
from sklearn import svm
sv=svm.SVC(kernel='linear')
sv.fit(x_train,y_train)
```

```
YPred=sv.predict(x_test)
```

```
print("accuracy : %.2f"% metrics.accuracy_score(YPred,y_test))
```

```
sv1=svm.SVC(kernel='poly', degree=8)
sv1.fit(x_train,y_train)
Ypred1=sv1.predict(x_test)
```

```
print("Accuracy : %.2f" % metrics.accuracy_score(Ypred1,y_test))
```

```
print(classification_report(Ypred1,y_test))
```

```
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```
rf.fit(x_train,y_train)
Ypredd=rf.predict(x_test)
```

```
from sklearn.metrics import accuracy_score
acc=accuracy_score(Ypredd,y_test)
```
