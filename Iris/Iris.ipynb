```
import numpy as np
import pandas as pd
```

```
d=pd.read_csv('Iris.csv')
```

```
d.head(10)
```

```
d.info
```

```
d['Species'].unique()
```

```
d['Species'].replace(str('Iris-setosa'),0,inplace=True)
d['Species'].replace(str('Iris-versicolor'),1,inplace=True)
d['Species'].replace(str('Iris-virginica'),2,inplace=True)
```

```
d.head(10)
```

```
d['Species'].unique()
```

```
d['Species'].astype(int)
```

```
d.dtypes
```

```
x=d[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=d['Species']
```

```
from sklearn.model_selection import train_test_split
x1,x2,y1,y2=train_test_split(x,y,test_size=0.3,random_state=42)
```

```
print(x1)
```

```
x1=np.array(x1).reshape(-1,4)
y1=np.array(y1).reshape(-1,1)
```

```
x1.shape
```

```
y1.shape
```

```
from sklearn import svm
s=svm.SVC(kernel='linear')
s.fit(x1,y1)
```

```
pred=s.predict(x2)
```

```
print(pred)
```

```
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y2, pred))
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y2, pred))
print(classification_report(y2, pred))
```

```
print("co-efficients : ",s.coef_)
print("Intercept : ",s.intercept_)
```
